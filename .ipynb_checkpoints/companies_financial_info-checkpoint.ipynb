```python
import requests

# Define your API key
api_key = 'crppfmhr01qsek0g9g1gcrppfmhr01qsek0g9g20'

# Step 1: Retrieve Semiconductor Companies from US Exchange
url = f'https://finnhub.io/api/v1/stock/symbol?exchange=US&token={api_key}'
response = requests.get(url)
companies = response.json()

# Step 2: Filter for Semiconductor Companies (description or name)
semiconductor_companies = [company for company in companies if 'semiconductor' in company['description'].lower()]

# Step 3: Fetch Financial Statements for each Semiconductor Company
financials = []
for company in semiconductor_companies:
    ticker = company['symbol']
    financial_url = f'https://finnhub.io/api/v1/stock/financials-reported?symbol={ticker}&token={api_key}'
    financial_response = requests.get(financial_url)
    financial_data = financial_response.json()
    financials.append({ticker: financial_data})

# Step 4: Print or process the financial data
for company_financials in financials:
    for ticker, financial_data in company_financials.items():
        print(f"Financial data for {ticker}:")
        print(financial_data)
        print("\n")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for the semiconductor sector on Finviz
url = 'https://finviz.com/screener.ashx?v=111&f=ind_semiconductor'

# Add a header to avoid being blocked by the website
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Get the webpage content
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table with ticker symbols
table = soup.find('table', class_='table-light')

# Extract all ticker symbols from the table
tickers = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    ticker = columns[1].text  # Second column is the ticker symbol
    tickers.append(ticker)

# Create a DataFrame
df = pd.DataFrame(tickers, columns=['Ticker'])

# Display the DataFrame
print(df)
```
