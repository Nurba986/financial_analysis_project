## 3. Column normalization


### 3.1. Convert columns 

```python
# Convert columns to datetime and categorical types.
# List of DataFrames to process
dataframes = {
    "balance_sheet_assets": balance_sheet_assets,
    "balance_sheet_equity": balance_sheet_equity,
    "balance_sheet_liabilities": balance_sheet_liabilities,
    "cash_flow_statement": cash_flow_statement,
    "eps_estimate": eps_estimate,
    "income_statement": income_statement,
    "sales_estimate": sales_estimate,
    "eps_history": eps_history
}

# Convert 'date' columns to datetime and 'act_symbol', 'period', 'period_end_date' to categorical
for name, df in dataframes.items():
    # Convert 'date' column to datetime if it exists in the DataFrame
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')  
    
    # Convert 'act_symbol', 'period', and 'period_end_date' columns to categorical if they exist in the DataFrame
    if 'act_symbol' in df.columns:
        df['act_symbol'] = df['act_symbol'].astype('category')
    if 'period' in df.columns:
        df['period'] = df['period'].astype('category')
    if 'period_end_date' in df.columns:
        df['period_end_date'] = df['period_end_date'].astype('category')

    # Display data types for verification
    print(f"\nData types for {name} after conversion:")
    print(df.dtypes)
```
